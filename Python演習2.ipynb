### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5
for i in range(1, size + 1):
 print('*' * i)
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
for i in range(size):
 print(' ' * (size - 1 - i) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
for i in range(size):
    for j in range(size):
        if j == i or j == size - 1 - i:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()   # ← ここ。for j の外
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
for i in range(1, size + 1):
  print(" " * (size - i) + "*" * (2 * i - 1))
````
